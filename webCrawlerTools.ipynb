## Crawler in Zap for list of tools

```python

# Import `requests`
import requests

url = 'https://www.zap.co.il/models.aspx?sog=h-handicrafttools&pageinfo=1'

# Make the request and check object type
r = requests.get(url)
type(r) 
```

```python

# Extract HTML from Response object and print
html = r.text
print(html)
```

```python

# Import BeautifulSoup from bs4
from bs4 import BeautifulSoup


# Create a BeautifulSoup object from the HTML
soup = BeautifulSoup(html, "html5lib")
type(soup)
```

```python

# Get soup title
soup.title
```

```python

# Get soup title as string
soup.title.string
```

```python

soup.findAll('div')
```

```python

tools=[]
for div in soup.findAll('div', {'class': 'SearchResultsMain '}):
        for div2 in div.findAll('div',{'class':'ProdInfoTitle'}):
            tools.append(div2.text.strip().replace(',',''))

print(tools)       
```

```python

def toolsList(soup):
    tools=[]
    for div in soup.findAll('div', {'class': 'SearchResultsMain '}):
        for div2 in div.findAll('div',{'class':'ProdInfoTitle'}):
            tools.append(div2.text.strip().replace(',',''))
    
    return tools
```

```python

print(list(set(toolsList(soup))))
```

```python

urlNumber = 'https://www.zap.co.il/models.aspx?sog=h-handicrafttools&pageinfo='

def run_pages(url,num_pages):
    tools=[]
    for i in range(1,num_pages):
        u=url+str(i)
        r = requests.get(u) 
        html = r.text
        soup = BeautifulSoup(html, "html5lib")
        newtools = toolsList(soup)
        tools.extend(newtools)
        print(i)
    return tools
```

```python
tools = run_pages(urlNumber,num_pages=10)
```

```python
tools
```

```python

import re

def clean_text(tools):
    tools2=[]
    for t in tools:
        t=re.sub("[^\na-z\u0590-\u05fe]", " ",t)
        t=(re.sub("[na-zA-Z]", "",t)).strip()
        tools2.append(t) 
    return tools2   

tools=clean_text(tools)
```

```python
tools
```

```python

tools = list(set(filter(None, tools)))
```

```python

import pandas as pd
pd.DataFrame(tools).to_csv('tools.csv' , index=False)
 
```
